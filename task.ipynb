# KNN
## Where KNN is used
  - similarity based prediction
  - Recommendation system
  - Pattern recognition
  - Image classification
  Examples:Movie recommendation,customer segmentation,disease prediction,handwriting recognition

## What is KNN(k-nearest neighbour)
  - works using similarity and distance
  - it doesnot learn equations like logistic regression or svm
  - works on both classification and regression
## How KNN works
  1. It will choose the k-value
  2. Find the nearest neighbour (if k=3, then 3 neighbours)
  3. Check majority class
  4. Assign predicted class
## Distance Metrics
  1. Euclidean distance-sqrt (x2-x2)^2 + (y2-y1)^2



```python
import math
x1 = 2
y1 = 3
x2 = 6
y2 = 7
distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
print("The distance between the points is:", distance)
```

1. Euclidean - continuous numerical data
2. Manhattan - Grid based distance
3. Minkowski - Generalized distance
4. Hamming - binary categorical data

## How to choose k-value
#### Small K (e.g., 1):
- Very sensitive to noise
- Can overfit
#### Large K:
- More stable
- Can underfit

Usually choose odd K (to avoid tie in classification)


```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
data = load_iris()

x = data.data
y = data.target

#split dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#Create KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=3, #Number of nearest neighbors
                                    metric='minkowski', #Distance metric
                                    p=2, #Power parameter for Minkowski distance (p=2 is equivalent to Euclidean distance
)

#Train the model
knn_classifier.fit(x_train, y_train)

#Predict
y_pred = knn_classifier.predict(x_test)

#Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
